```python
import numpy as np
import pandas as pd
import datetime
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

def get_dataset():
    X = pd.read_csv('clean/X_train.csv')
    X = X.drop('index', axis = 1)
    bad = (X['Married(1/0)'] == -1).values
    X.loc[bad, 'Married(1/0)'] = 1
    y = pd.read_csv('clean/y_train.csv')
    df = pd.concat([X, y], axis = 1)
    return df

def get_test():
    X_test = pd.read_csv('clean/X_test.csv')
    X_test = X_test.drop('index', axis = 1)
    return X_test

from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from xgboost import XGBClassifier
import xgboost
```

```python
def feat_eng(X_train, X_test):
    data = pd.concat([X_train, X_test]).reset_index()
    imp_cols = ['Role', 'Invested in Real Estate', 'Extra Hours',
       'Invested in Stock Market', 'Favourite Sport', 'Gender',
       'Languages Known', 'Sport Knowledge (in XP)'
       , 'Trades Done', 'Married(1/0)',
       'Level of Education', 'Pay', 'ID Proof', 'Most Productive Hour',
       'Tax Category', 'Income Category', 'Employer', 'Tax Paid']
    data = data[imp_cols]
    ntrain = X_train.shape[0]
    return data[:ntrain], data[ntrain:]

for _ in range(4):
    
    model = RandomForestClassifier(n_estimators=121, max_depth=18)
    df = get_dataset()
    X = df.drop('Occupation', axis = 1)
    y = df['Occupation']
    test = get_test()
        
#     X, test = feat_eng(X, test)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    score = metrics.accuracy_score(preds, y_test)
    print(classification_report(preds, y_test))
    print(f"Score = {score}")
    if score >= 0.44:
        print('FOUND\n\n')
        answers = model.predict(test)
        make_sub('RandomForest', score, answers)
        
```
