```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,5))
sns.displot(df['SalePrice'], kde = False)
```

```python
plt.figure(figsize=(12,5))
sns.displot(np.log(df['SalePrice']), kde = False)
```

```python
corr = df.select_dtypes(
    include = ['float64', 'int64']
).iloc[:, 1:].corr()
sns.set(font_scale = 1)
sns.heatmap(corr, vmax = 1, square = True)
```

```python
corr.head()
```

```python
corr_list = corr['SalePrice'].sort_values(axis = 0, ascending = False)
corr_list = corr_list.iloc[1:]
corr_list.head()
```

```python
plt.figure(figsize=(18,8))
for i in range(6):
    ii = '23'+str(i+1)
    plt.subplot(ii)
    feature = corr_list.index.values[i]
    plt.scatter(df[feature], df['SalePrice'], facecolors='none',edgecolors='k',s = 75)
    sns.regplot(x = feature, y = 'SalePrice', data = df,scatter=False, color = 'Blue')
    ax=plt.gca() 
    ax.set_ylim([0,800000])
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'Neighborhood', y = 'SalePrice',  data = df)
xt = plt.xticks(rotation=45)
```
